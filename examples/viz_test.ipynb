```python
from symop.core.operators import ModeOp
from symop.core.monomial import Monomial
from symop.core.terms import KetTerm, DensityTerm
from symop.ccr.algebra.ket.poly import KetPoly
from symop.ccr.algebra.op.poly import OpPoly
from symop.ccr.algebra.density.poly import DensityPoly
from symop.modes.envelopes import GaussianEnvelope
from symop.modes.labels import ModeLabel, Path, Polarization
from symop.polynomial.state import KetPolyState
from symop.polynomial.state import DensityPolyState
import symop.viz as VI
```

```python
env_1 = GaussianEnvelope(omega0=10, sigma=2, tau=0)
env_2 = GaussianEnvelope(omega0=5, sigma=2, tau=0.5)
path_1 = Path(name="A")
pol_1 = Polarization.D()

mode_label_1 = ModeLabel(path=path_1, envelope=env_1, polarization=pol_1)
mode_label_2 = ModeLabel(path=path_1, envelope=env_2, polarization=pol_1)
mode_op_1 = ModeOp(label=mode_label_1)
mode_op_2 = ModeOp(label=mode_label_2)
a_dag = mode_op_1.create
a = mode_op_1.ann
b_dag = mode_op_2.create
b = mode_op_2.ann

monomial_1 = Monomial(creators=(a_dag, a_dag))
ket_term = KetTerm(coeff=1.0, monomial=monomial_1)

monomial_2 = Monomial(creators=(b_dag, b_dag), annihilators=(a, b))
ket_poly = KetPoly(terms=(ket_term,))
ket_term_2 = KetTerm(coeff=1.0, monomial=monomial_2)
density_poly = DensityPoly.pure(ket_poly)
op_poly_1 = OpPoly.n(mode_op_2)
VI.display(a_dag)
VI.display(a)
VI.display(monomial_1)
#_ = VI.plot(monomial_1)
print("KET TERM")
VI.display(ket_term)
print("KET POLY")
VI.display(ket_poly)
VI.display(KetTerm(coeff=1.0, monomial=Monomial.identity()))
print("Op POLY")
VI.display(op_poly_1)
print("Density POLY")
VI.display(density_poly)
```

```python
ket_state = KetPolyState(ket=ket_poly)
VI.display(ket_state)
density_state = DensityPolyState.pure(ket_state)
VI.display(density_state)
```
