```python
import sys
from math import sqrt
from symop_proto.core.monomial import Monomial
from symop_proto.core.operators import ModeOp, LadderOp
from symop_proto.envelopes import GaussianEnvelope
from symop_proto.labels import ModeLabel, PolarizationLabel, PathLabel
from symop_proto.algebra.polynomial import KetPoly, DensityPoly
from symop_proto.algebra.operator_polynomial import OpPoly
from symop_proto.state import KetPolyState, DensityPolyState
from symop_proto.core.terms import DensityTerm
```

```python
env1 = GaussianEnvelope(omega0=0, sigma=1.0, tau=0)
label1 = ModeLabel(pol=PolarizationLabel.H(), path=PathLabel(name="A"))
env2 = GaussianEnvelope(omega0=0, sigma=3.0, tau=0)
label2 = ModeLabel(pol=PolarizationLabel.H(), path=PathLabel(name="A"))
m1 = ModeOp(env=env1, label=label1).with_label("o")
m2 = ModeOp(env=env2, label=label2).with_label("e")

psi = KetPolyState.vacuum().with_label("a")
display(psi.normalized())
psi = m1 @ ( m1 @ (m1 @ psi))
display(psi.normalized())

exp = OpPoly.from_words(words=((m1.create,m1.ann),))
print(psi.expect(exp))
display(psi)
print(psi.label)
display(exp)
```

```python
q2 = OpPoly.q2(m1)
p2 = OpPoly.p2(m1)
print(KetPolyState.vacuum().expect(q2))
print(KetPolyState.vacuum().expect(p2))
```

```python
def make_mode(
    path: str = "A",
    *,
    omega: float = 1.0,
    sigma: float = 0.3,
    tau: float = 0.0,
    phi: float = 0.0,
) -> ModeOp:
    env = GaussianEnvelope(omega0=omega, sigma=sigma, tau=tau, phi0=phi)
    label = ModeLabel(PathLabel(path), PolarizationLabel.H())
    return ModeOp(env=env, label=label)
m = make_mode("A")
rho0 = DensityPoly(terms=(DensityTerm(1.0, Monomial(), Monomial()),))
display(rho0)
# Left apply a^dag then a
rhoL = rho0.apply_left((m.create, m.ann))
sigsL = {t.left.signature for t in rhoL.terms}
rhoR = rho0.apply_right((m.ann,))
display(rhoL)
display(rhoR)
```
