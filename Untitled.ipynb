```python


from symop_proto.core.operators import ModeOp, LadderOp
from symop_proto.envelopes import GaussianEnvelope
from symop_proto.labels import ModeLabel, PolarizationLabel, PathLabel

from symop_proto.algebra.polynomial import KetPoly, DensityPoly
from symop_proto.algebra.operator_polynomial import OpPoly
from symop_proto.state import KetPolyState, DensityPolyState

# --- tiny helpers ---
def show_expect(psi_state: KetPolyState, words, label=""):
    """words: list[tuple[LadderOp,...]] representing an OpPoly."""
    O = OpPoly.from_words(words=words)
    raw = DensityPoly.pure(psi_state.ket)
    den = raw.trace()
    num = sum(t.coeff * raw.apply_right(t.ops).trace() for t in O.terms)
    print(f"{label:10s}  Tr(ρO)={num:.6g},  Tr(ρ)={den:.6g},  ⟨O⟩={(num/den):.6g}")

def check_close(x, y, tol=1e-10, label=""):
    ok = abs(x - y) <= tol
    print(f"{label:20s} {'OK' if ok else 'FAIL'}  (got {x}, want {y})")
    return ok

```

```python
env = GaussianEnvelope(omega0=0.0, sigma=1.0, tau=0.0)
label = ModeLabel(pol=PolarizationLabel.H(), path=PathLabel(name="A"))
m = ModeOp(env=env, label=label)

adag, a = m.create, m.ann

# OpPolys for common words
N   = OpPoly.from_words(words=[(adag, a)])   # number operator a† a
N1  = OpPoly.from_words(words=[(a, adag)])   # a a† (gives n+1 on |n>)

```

```python
# |0>
psi0 = KetPolyState.vacuum()
print("Vacuum normalized? ", psi0.is_normalized())

# a† |0> (single photon)
psi1 = adag @ psi0
print("Single-photon norm2:", psi1.norm2)
print("⟨N⟩ on |1> (should be 1):", psi1.expect(N))
print("⟨a a†⟩ on |1> (should be 2):", psi1.expect(N1))
```

```python
def n_photon_state(n: int) -> KetPolyState:
    psi = KetPolyState.vacuum()
    for _ in range(n):
        psi = adag @ psi
    return psi

for n in (0, 1, 2, 3):
    psi = n_photon_state(n)
    rho = DensityPoly.pure(psi.ket)
    tr  = rho.trace()            # should be n!
    expN = psi.expect(N)         # should be n (normalized by default)
    expN1= psi.expect(N1)        # should be n+1

    print(f"\n--- |n> with n={n} ---")
    print("norm2 (n!)    :", psi.norm2)
    print("Tr(rho)       :", tr)
    print("⟨a†a⟩ (want n) :", expN)
    print("⟨a a†⟩ (want n+1):", expN1)
```
